```python
evclass=pd.read_csv('Vehicle Class - All.csv')
evclass.head()
```

```python
evclass.shape
```

```python
evclass.dtypes
```

```python
evclass['Total Registration']=evclass['Total Registration'].str.replace(',','')
```

```python
evclass.head()
```

```python
evclass['Total Registration']=evclass['Total Registration'].astype(int)
```

```python
evclass.dtypes
```

```python
evclass['Vehicle Class'].unique()
```

```python
evclass['Vehicle Class']=evclass['Vehicle Class'].str.replace('FOUR WHEELER (INVALID CARRIAGE)','FOUR WHEELER')
evclass['Vehicle Class']=evclass['Vehicle Class'].str.replace('THREE WHEELER(NT)','THREE WHEELER')
evclass['Vehicle Class']=evclass['Vehicle Class'].str.replace('THREE WHEELER(T)','THREE WHEELER')
evclass['Vehicle Class']=evclass['Vehicle Class'].str.replace('TWO WHEELER(NT)','TWO WHEELER')
evclass['Vehicle Class']=evclass['Vehicle Class'].str.replace('TWO WHEELER(INVALID CARRIAGE)','TWO WHEELER')
evclass['Vehicle Class']=evclass['Vehicle Class'].str.replace('TWO WHEELER(T)','TWO WHEELER')
```

```python
evclass.head()
```

```python
evclass=evclass[evclass['Vehicle Class']!='OTHER THAN MENTIONED ABOVE']
```

```python
evclass['Vehicle Class'].unique()
```

```python
evclass['Vehicle Class']=evclass['Vehicle Class'].str.replace('TWO WHEELER(INVALID CARRIAGE)','TWO WHEELER')
```

```python
evclass['Vehicle Class'].unique()
```

```python
plt.figure(figsize=(35,28))
sns.barplot(x=evclass['Vehicle Class'],y=evclass['Total Registration'],data=evclass)
plt.show()
```

```python
evsales=pd.read_csv('ev_sales_by_makers_and_cat_15-24.csv')
evsales.head()
```

```python
evs=evsales.melt(id_vars=['Cat','Maker'],
                 var_name='Year',
                 value_name='Count')
evs.head()
```

```python
cat_year_data=evs.groupby(['Cat', 'Year'])['Count'].sum().reset_index()
cat_year_data.head()
```

```python
cat_year_data['Cat'].unique()
```

```python
sns.barplot(x=cat_year_data[cat_year_data['Cat']=='2W']['Year'],y=cat_year_data['Count'])
plt.title('Year Vs Count of 2W')
plt.show()
```

```python
sns.barplot(x=cat_year_data[cat_year_data['Cat']=='3W']['Year'],y=cat_year_data['Count'])
plt.title('Year Vs Count of 3W')
plt.show()
```

```python
sns.barplot(x=cat_year_data[cat_year_data['Cat']=='LMV']['Year'],y=cat_year_data['Count'])
plt.title('Year Vs Count of LMV')
plt.show()
```

```python
sns.barplot(x=cat_year_data[cat_year_data['Cat']=='MMV']['Year'],y=cat_year_data['Count'])
plt.title('Year Vs Count of MMV')
plt.show()
```
